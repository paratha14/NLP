```python
import nltk
import spacy
```

```python
from nltk.stem import PorterStemmer
stemmer= PorterStemmer()
```

```python
words = [
    "run",
    "apple",
    "quick",
    "slowly",
    "jump",
    "table",
    "bright",
    "carefully",
    "sing",
    "book",
    "ability"
]
```

```python
for word in words:
  print(word, ":", stemmer.stem(word))
```

```python
nlp= spacy.load("en_core_web_sm")
doc= nlp("run apple quick slowly jump table bright carefully sing book ability")
```

```python
for token in doc:
  print(token, ": ", token.lemma_)
```

Customizing the lemmatized words



```python
nlp.pipe_names
```

```python
#normal or default lemmatization
doc= nlp("Bro, you wanna go? Brah, don't say no! I am exhausted")

for token in doc:
  print(token,"| ", token.lemma_)
```

```python
#as bro and brah means brother hence:
#using attribute ruler

ar= nlp.get_pipe('attribute_ruler')
ar.add([[{'TEXT':'Bro'}],[{'TEXT':'Bruh'}]], {'LEMMA':'Brother'})

doc= nlp("Bro, you wanna go? Brah, don't say no! I am exhausted")
for token in doc:
  print(token,'|',token.lemma_)
```
