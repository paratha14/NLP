```python
import spacy
```

```python
nlp= spacy.load('en_core_web_sm')
```

```python
doc= nlp("Parker flew to NewJersey yesterday. He carried webshooters with him")
for token in doc:
  print(token,'|', token.pos_, '|', spacy.explain(token.pos_), '|', token.tag_, '|', spacy.explain(token.tag_))
```

```python
text= '''Abhishek Sharma and Ishan Kishan are getting ready on the sidelines as the USA players take the field. Big test for Saurabh with the new ball bowling to these two batters. He will love this challenge.
Teams are out for the anthems.
India to score 70+ in the powerplay: The ball flies at the Wankhede so it won't be surprising to see India score in excess of 70 in the first six overs. Expect the total to cross 80 if Abhishek Sharma and Ishan Kishan bat for more than 4 overs.
Rohit Sharma: We had a lot of memories. Of course, when we lifted the trophy, that was the best memory. But the entire journey that we had from day one till the last day of the tournament was itself a memory for us, something that we can always cherish. We worked really, really hard for that trophy. And from day one, since the day we arrived in the US and the day we left Barbados, everything was memorable for us. It was a wonderful experience. Look, something that I never expected so early in my career, to be an ICC ambassador. You know, it’s a big honour. Holding the trophy back again and putting it back in play feels great. Lifting the trophy was a wonderful feeling, but again, to put it back into play and look at these boys going and working hard for that trophy again will be an interesting thing, which I’m pretty sure everyone’s looking forward to. We want to see some really good cricket being played in this tournament, which is what ICC brings. It brings the best of cricket on display. Every team who are playing, they bring the best cricket out there, and it’ll be entertaining for all the spectators to watch. '''
```

```python
doc2= nlp(text)
for token in doc2:
  print(token,'|', token.pos_)
```

Removing all punctuations and spaces present.

```python
filtered=[]
for token in doc2:
  if token.pos_ not in ['SPACE', 'PUNCT']:
    filtered.append(token)
for t in filtered:
  print(t,'|', t.pos_)
```

Extracting count of each of the POS

```python
count= doc2.count_by(spacy.attrs.POS)
type(count)
```

```python
for k, v in count.items():
  print(doc2.vocab[k].text, '|', v)
```
