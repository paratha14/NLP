```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
corpus = [
    "The quick brown fox jumps over the lazy dog.",
    "Machine learning models require extensive data preprocessing before training.",
    "In Lucknow, street food vendors sell spicy chaat during evening markets.",
    "She trained a neural network using Python and Keras on Google Colab.",
    "Elephants roam freely in the dense forests of Uttar Pradesh.",
    "Hyperparameter tuning improved the accuracy of the random forest classifier.",
    "The startup pitched an AI app for sports analytics at the hackathon.",
    "Natural language processing handles sarcasm detection with challenging datasets.",
    "He practiced mixed martial arts three times a week in the local gym.",
    "Data visualization reveals hidden patterns in large customer transaction logs."
]
```

```python
v= TfidfVectorizer()
transformed= v.fit_transform(corpus)
print(v.vocabulary_)
```

```python
all_feature_names= v.get_feature_names_out()
```

```python
for word in all_feature_names:
    indx= v.vocabulary_.get(word)
    print(f"word: {word} idf score: {v.idf_[indx]}")
```

```python
target={"Household":0, "Books":1, "Electronics":2, "Clothing & Accessories": 3}
df['label_num']= df.label.map(target)
df.head()
```

```python
df.isna().any()
```

```python
df.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(
    df.text,
    df.label_num,
    test_size=0.2,
    random_state=2022,
    stratify=df.label_num
)
```

```python
print("Shape of X_train: ", X_train.shape)
print("Shape of X_test: ", X_test.shape)
```

```python
print("Shape of y_train: ", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

##1.KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report


knn_pipe= Pipeline([
    ('tf-idf', TfidfVectorizer()),
    ('knn', KNeighborsClassifier())
    ])

knn_pipe.fit(X_train, y_train)

yp_knn= knn_pipe.predict(X_test)
print(classification_report(y_test, yp_knn))
```

##2. NaiveBayes

```python
from sklearn.naive_bayes import MultinomialNB
nb_pipe= Pipeline([
    ('tf-idf', TfidfVectorizer()),
    ('nb', MultinomialNB())
    ])

nb_pipe.fit(X_train, y_train)

yp_nb= nb_pipe.predict(X_test)
print(classification_report(y_test, yp_nb))
```

##3. RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
f_pipe= Pipeline([
    ('tf-idf', TfidfVectorizer()),
    ('forest', RandomForestClassifier())
    ])

f_pipe.fit(X_train, y_train)

yp_f= f_pipe.predict(X_test)
print(classification_report(y_test, yp_f))
```
