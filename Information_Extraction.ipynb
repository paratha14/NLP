```python
import re
```

```python
text= '''
Booking Confirmed!
Your ticket has been successfully booked.

Booking ID: BK98234176
Name: Rahul Sharma
Travel Date: 18 Feb 2026
Route: New Delhi → Mumbai
Seat: A12 (Coach S2)

Contact: +91 98765 43210 | rahul.sharma@gmail.com

Amount Paid: ₹2,450 (UPI)
Txn ID: TXN564839201

Have a safe journey!
'''
```

```python
confirmation= re.search("Confirmed", text)
```

```python
booking_number= re.findall("Booking ID:\s*([A-Za-z0-9]+)", text)
booking_number
```

```python
name= re.findall("Name:\s*([A-Za-z]+\s*[A-za-z]+)", text)
name
```

```python
travel_date= re.findall('Travel\s*Date\s*:\s*([0-9]+\s*[A-Za-z]+\s*[0-9]+)', text)
travel_date
```

```python
route= re.findall('Route:\s*[A-za-z]+\s*([A-Za-z]+\s*→\s*[A-Za-z]+)', text)
route
```

```python
seat= re.findall('Seat:\s*([A-Z0-9]+\s*\([A-Za-z]+\s*[A-Z0-9]+\))', text)
seat
```

```python
phno= re.findall('(\+\d{2}\s*\d{5}\s*\d{5})', text)
phno
```

```python
print(f"Confirmation Status: {confirmation[0]}")
print(f"Booking Number: {booking_number[0]}")
print(f"Name: {name[0]}")
print(f"Travel Date: {travel_date[0]}")
print(f"Route: {route[0]}")
print(f"Seat: {seat[0]}")
print(f"Phone Number: {phno[0]}")
```
