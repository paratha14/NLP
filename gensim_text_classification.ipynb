```python
word_vec= api.load('glove-wiki-gigaword-100')
```

```python
import spacy
!python -m spacy download en_core_web_lg
nlp= spacy.load("en_core_web_lg")
```

```python
def preprocessing_vectorizing(text):
    doc= nlp(text)
    filtered_tokens=[]

    for token in doc:
        if token.is_punct or token.is_stop:
            continue
        else:
            filtered_tokens.append(token.lemma_)

    return word_vec.get_mean_vector(filtered_tokens)
```

```python
preprocessing_vectorizing("Hi how are you!")
```

```python
df= df.head(1200)
```

```python
print(df.shape)
df.head()
```

```python
df['vector_text']= df['text'].apply(lambda x: preprocessing_vectorizing(x))
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(
    df.vector_text.values,
    df.label,
    test_size=0.15,
    random_state=2019
)
```

```python
import numpy as np
X_train_2d= np.stack(X_train)
X_test_2d= np.stack(X_test)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler
mn= MinMaxScaler()
X_train_2d= mn.fit_transform(X_train_2d)
model= MultinomialNB()
model.fit(X_train_2d, y_train)
```

```python
X_test_2d= mn.transform(X_test_2d)
model.score(X_test_2d, y_test)
```

```python
from sklearn.metrics import classification_report
yp= model.predict(X_test_2d)
print(classification_report(y_test, yp))
```
