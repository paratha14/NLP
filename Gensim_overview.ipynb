```python
import gensim.downloader as api
```

```python
model = api.load('glove-wiki-gigaword-100')
```

```python
model.similarity("great", "good")
```

```python
model.similarity("profit", "gain")
```

```python
model.most_similar("dogs")
```

```python
model.most_similar(positive=['king', 'woman'], negative=['man'])
```

```python
model.most_similar(positive=['france', 'berlin'], negative=['paris'])[0]
```

```python
model.doesnt_match(['breakfast', 'cereal', 'dinner'])
```
