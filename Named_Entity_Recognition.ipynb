```python
sentence= "11.3: Keshav Maharaj to Rachin Ravindra, Shorter and outside off, Rachin hits it along the ground to long off for a run."
```

```python
doc = nlp(sentence)
```

```python
for i in doc:
  if(i.pos_ != "PUNCT"):
    print(f"{i} | {i.tag_} | {i.lemma_} | {i.pos_}")
```

```python
for ents in doc.ents:
  print(f"{ents.text} | {ents.label_} | {spacy.explain(ents.label_)}" )
```

```python
from spacy import displacy
displacy.render(doc, style="ent")
```

```python
doc3= nlp("Michael Bloomberg founded Bloomberg in 1982")
```

```python
from spacy.tokens import Span
s1= Span(doc3,3,4, label="ORG")

doc3.set_ents([s1], default="unmodified")
```

```python
for ents in doc3.ents:
  print(f"{ents.text} | {ents.label_} | {spacy.explain(ents.label_)}" )
```
