```python
import nltk
import spacy
```

```python
from nltk.stem import PorterStemmer
stemmer= PorterStemmer()
```

```python
words = [
    "run",
    "apple",
    "quick",
    "slowly",
    "jump",
    "table",
    "bright",
    "carefully",
    "sing",
    "book",
    "ability"
]
```

```python
for word in words:
  print(word, ":", stemmer.stem(word))
```

```python
nlp= spacy.load("en_core_web_sm")
doc= nlp("run apple quick slowly jump table bright carefully sing book ability")
```

```python
for token in doc:
  print(token, ": ", token.lemma_)
```
