```python
import spacy
```

```python
nlp_spacy= spacy.load("en_core_web_sm")
```

```python
text= '''Your order has been confirmed successfully. It will be delivered to your address within 3–5 business days. For any queries, contact support at support@example.com
 or call 9876543210.'''
```

```python
docs= nlp_spacy(text)
```

```python
for idx,sentence in enumerate(docs.sents):
  print(f"Sentence {idx+1}: {sentence}")
print("\nwords are:\n")
for sentence in docs.sents:
  for words in sentence:
    print(words)





```
